```python
class node:
    def __init__(self,val,left=None,right=None,parent=None):
        self.val = val
        self.left = left
        self.right = right
        self.parent = parent
        pass
    pass


class BST:
    def __init__(self,nodes=[]):
        self.tree = None
        self.insert(*nodes)
        pass
    
    def search(self,val):
        curr_node = self.tree
        while curr_node:
            curr_val = curr_node.val
            if curr_val == val:
                return curr_node
            elif curr_val>val:
                curr_node = curr_node.left
            else:
                curr_node = curr_node.right
        return
    
    
    def insert(self,*args):
        for n in args:
            if self.tree:
                this_node = node(n)
                curr_head = self.tree
                while curr_head:
                    curr_val = curr_head.val
                    if curr_val>=n:
                        prev_head = curr_head
                        curr_head = prev_head.left
                    else:
                        prev_head = curr_head
                        curr_head = prev_head.right
                if prev_head.val>=n:
                    prev_head.left = this_node
                else:
                    prev_head.right = this_node
                this_node.parent = prev_head
            else:
                self.tree = node(n)
        pass
    
    def delete(self,key):
        
        curr_node = self.search(key)
        
        
        #k has no children
        if not curr_node.left and not curr_node.right:
            parent = curr_node.parent
            if parent:
                if parent.left == curr_node:
                    parent.left = None
                else:
                    parent.right = None
            else: #must be the root node and the tree does not have anything else
                self.tree = None
        
        #k has one child
        elif not curr_node.left and curr_node.right:
            parent = curr_node.parent
            if parent:
                if parent.left == curr_node:
                    parent.left = curr_node.right
                else:
                    parent.right = curr_node.right
            else: #must be the root node
                self.tree = curr_node.right
        
        elif not curr_node.right and curr_node.left:
            parent = curr_node.parent
            if parent:
                if parent.left == curr_node:
                    parent.left = curr_node.left
                else:
                    parent.right = curr_node.left
            else: #must be the root node
                self.tree = curr_node.left
                
        #k has 2 child
        else:
            pred = self.find_predecessor(curr_node.val)
            if pred.left:
                pred.parent.right = pred.left
            else:
                pred.parent.right = None
            #pred changed to curr
            curr_node.val = pred.val
                
            
    
    def find_predecessor(self,key):
        curr_node = self.search(key)
        if curr_node.left:
            pred = curr_node.left
            while pred:
                prev_pred = pred
                pred = prev_pred.right
            return prev_pred
    
    def _print_tree_preorder(self,node):
        if node:
            print(node.val)
            self._print_tree_preorder(node.left)
            self._print_tree_preorder(node.right)
        
    def print_tree(self):
        self._print_tree_preorder(self.tree)
```
